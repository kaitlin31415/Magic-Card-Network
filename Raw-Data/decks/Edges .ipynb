```python
with open("edges.csv", "w") as f:
    files = [f for f in os.listdir('./raw_decks')]
    for f_1 in files:
        filename = str(f_1)
        print(filename)
        df = pd.read_csv('./raw_decks/' + filename)
        df = pd.merge(df, cards_metadata, left_on="Name", right_on="name")
        commanders = df[df['Commander'] == True]
        deck = df[df['Commander'] != True]
        for commander in commanders['id'].unique():
            print(commander)
            deck['commander'] = commander
            for card in deck['id'].unique():
                f.write("{},{}\n".format(card, commander))
```
