# Nodes Information

```python
import pandas as pd
```

```python
import json
 
# Opening JSON file
with open('card_info.json', encoding="utf-8") as json_file:
    data = json.load(json_file)
```

```python
dataFrameHistorical = pd.DataFrame() 
dfItem = pd.DataFrame.from_records(data)               
dataFrameHistorical = dataFrameHistorical.append(dfItem, ignore_index=True)  
```

```python
 pd.DataFrame.from_dict(data).describe()
```

```python
dataFrameHistorical = pd.DataFrame() 

for item in data['cards']:
    dfItem = pd.DataFrame.from_dict(item, orient='index')
    dfItem = dfItem.transpose()
    dataFrameHistorical = dataFrameHistorical.append(dfItem, ignore_index=True)  
```

```python
dataFrameHistorical.to_csv("All_card_info_csv.csv")
```
